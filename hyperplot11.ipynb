```python
from dataclasses import dataclass

@dataclass
class PlotParams:
    measurement: str
    dataset: int
    network: str
    metric: str
    N: int
```

```python
def compute_curve(plot_params: PlotParams) -> pd.DataFrame:
    rows = (
        df
            .loc[df.Measurement == plot_params.measurement]
            .loc[df.Dataset == plot_params.dataset]
            .loc[df.Network == plot_params.network]
            .loc[df.N == plot_params.N]
    )

    if plot_params.measurement == 'loss':
        values = pd.DataFrame([report[row.Key] for row in rows.itertuples()])
    else:
        values = pd.DataFrame([report[row.Key][plot_params.metric] for row in rows.itertuples()])

    result = values.quantile([0.25, 0.75]).transpose()
    result['mean'] = values.mean()
    return result

pd.concat([compute_curve(
    PlotParams(
        measurement='evaluate',
        dataset=1,
        network='RegularNetwork',
        metric='accuracy',
        N=0,
    )
), compute_curve(
    PlotParams(
        measurement='evaluate',
        dataset=3,
        network='AugmentedReLUNetwork',
        metric='mse',
        N=6,
    )
)], axis=1)

```

```python
import matplotlib.pyplot as plt

def plot_curve(
        curve: pd.DataFrame,
        title: str,
        xlabel: str,
        ylabel: str,
    ):
    plt.plot(curve.index, curve['mean'], label='Mean', color='blue')
    plt.fill_between(curve.index, curve[0.25], curve[0.75], color='gray', alpha=0.5, label='0.25 to 0.75 Quantiles')
    plt.xlabel(xlabel)
    plt.ylabel(ylabel)
    plt.title(title)
    plt.legend()
    plt.show()

plot_params=PlotParams(
    measurement='loss',
    dataset=1,
    network='AugmentedReLUNetwork',
    metric='loss',
    N=2,
)

plot_curve(
    curve=compute_curve(plot_params=plot_params),
    title=f'{plot_params.measurement}: {plot_params.network}; N = {plot_params.N} dataset = {plot_params.dataset}',
    xlabel='Iteration',
    ylabel='Loss',
)
```
